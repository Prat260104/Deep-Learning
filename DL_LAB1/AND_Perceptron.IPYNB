```python
def pereceptron(x1,x2):
    w1,w2=1,1
    bias=-1.5
    z=w1*x1+w2*x2+bias
    if z>=0:
        return 1
    else:
        return 0

inputs = [(0,0),(0,1),(1,0),(1,1)]
for x in inputs:
    print(x,"->",pereceptron(x[0],x[1]))
```

```python
# OR Gate Perceptron
def OR_perceptron(x1,x2):
    w1,w2=1,1
    bias=-0.5
    z=w1*x1+w2*x2+bias
    if z>=0:
        return 1
    else:
        return 0

print("OR Gate:")
inputs = [(0,0),(0,1),(1,0),(1,1)]
for x in inputs:
    print(x,"->",OR_perceptron(x[0],x[1]))
```

```python
# NAND Gate Perceptron
def NAND_perceptron(x1,x2):
    w1,w2=-1,-1
    bias=1.5
    z=w1*x1+w2*x2+bias
    if z>=0:
        return 1
    else:
        return 0

print("NAND Gate:")
inputs = [(0,0),(0,1),(1,0),(1,1)]
for x in inputs:
    print(x,"->",NAND_perceptron(x[0],x[1]))
```

```python
# NOR Gate Perceptron
def NOR_perceptron(x1,x2):
    w1,w2=-1,-1
    bias=0.5
    z=w1*x1+w2*x2+bias
    if z>=0:
        return 1
    else:
        return 0

print("NOR Gate:")
inputs = [(0,0),(0,1),(1,0),(1,1)]
for x in inputs:
    print(x,"->",NOR_perceptron(x[0],x[1]))
```

```python
# NOT Gate Perceptron
def NOT_perceptron(x):
    w=-1
    bias=0.5
    z=w*x+bias
    if z>=0:
        return 1
    else:
        return 0

print("NOT Gate:")
inputs = [0,1]
for x in inputs:
    print(x,"->",NOT_perceptron(x))
```
